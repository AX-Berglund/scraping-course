#### Fetch the movies

```python
listOfMovies = soup.find('ul', class_='ipc-metadata-list')
```

```python
movieInList = listOfMovies.find_all('li', class_="ipc-metadata-list-summary-item sc-4929eaf6-0 DLYcv cli-parent")
```

#### Title
So to find the title, we search for it in the webpage. Note that it is in a `h3` tag. We use the `find` method to find the first `h3` tag and then get the text from it.

```python
movieTitle = movieInList[15].find('h3', class_ = "ipc-title__text")
print(movieTitle.text)
```

#### `find`returns the first element

```python
movieReleaseYear = movieInList[15].find('span', class_="sc-300a8231-7 eaXxft cli-title-metadata-item").text
print(movieReleaseYear)
```

#### `findAll` returns all the elements

```python
movieDuration = movieInList[15].findAll('span', class_="sc-300a8231-7 eaXxft cli-title-metadata-item")
print(movieDuration[1].text)
```

#### Get the url

```python
url = movieInList[15].find('a', class_="ipc-title-link-wrapper").get('href')
print(url)
```

#### Rating

```python
rating = movieInList[15].find('span', class_="ipc-rating-star--rating").text
print(rating)
```

```python
info = []
for movie in listOfMovies:
    # get information 
    title = movie.find('h3',class_='ipc-title__text').get_text()
    url = movie.find('a',class_='ipc-title-link-wrapper').get('href')
    year = movie.find('span',class_='cli-title-metadata-item').get_text()
    duration = movie.findAll('span',class_='cli-title-metadata-item')[1].get_text()
    age_group =movie.findAll('span',class_='cli-title-metadata-item')[2].get_text()
    rating = movie.find('span',class_='ipc-rating-star').get('aria-label')
    nb_ratings = movie.find('span',class_='ipc-rating-star--voteCount').get_text()
    nb_ratings = nb_ratings[nb_ratings.find("(")+1:nb_ratings.find(")")]
    # save information in dictionary : note, you don't need to do all this in several lines.
    data = dict()
    data['title'] = title
    data['year'] = year
    data['duration'] = duration
    data['age_group'] = age_group
    data['rating'] = rating
    data['nb_ratings'] = nb_ratings
    info.append(data)
print(info)
```
